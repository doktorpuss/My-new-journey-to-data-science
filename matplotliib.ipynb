```python
import matplotlib.pyplot as plt
import numpy as np
```

# **1. Tạo giá trị để biểu diễn**

```python
x = np.linspace(0,10,1000)
sin_x = np.sin(x)
cos_x = np.cos(x)
rev_parabol = x**2

rng = np.random.RandomState(0)
scatter_x = rng.randn(50)
scatter_y = rng.randn(50)
sizes = scatter_x = 10*rng.randn(50)
colors = rng.randn(50)

```

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig, ((line,scatter),(histogram,bar)) = plt.subplots(nrows=2,
                                                     ncols=2,
                                                     figsize=(10,5))

line.plot(sin_x,color='b', linestyle="--", label="sin(x)",linewidth = 2);
line.plot(cos_x, marker = 's');


scatter.scatter(scatter_x,scatter_y,c=colors, cmap = 'viridis',s=sizes);
scatter.xlim([0,20]);
```
